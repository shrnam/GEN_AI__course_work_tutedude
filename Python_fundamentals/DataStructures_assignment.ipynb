#Task 2 - Categories(Sets)

```python
#since the updated products list contains 8 elements, respective categories list also contains 8 elements

categories = ['electronics','clothing','furniture','food','medicine','food','electronics','clothing']
```

```python
#categories like clothing, electronics, food appear more than once but upon set() the duplicates are removed

categories_set = set(categories)
print(categories_set)
```

```python
categories_set.add('food') #added a category that is already existing, shows duplicates ignored
categories_set.add('movie') #added a new category that is not present
print(categories_set)
```

```python
print('food' in categories_set)
```

```python
print('toy' in categories_set)
```

```python
print(len(categories_set)) #no of unique categories
```

#Task 3 - Product Pricing (Dictionaries)

```python
print(products)
```

```python
keys = products
values = [120000,600,200,450,20,50,200000,300] #creating prices for each product in list
price_dict = dict(zip(keys,values))
print(price_dict)
```

```python
#adding new product with price to dict
price_dict['table'] = 1000
```

```python
#update the price of chair
price_dict['chair'] = 250
```

```python
#remove the product macbook
price_dict.pop('macbook',None)
```

```python
price_dict.pop('car',None)
```

```python
print(price_dict)
```

```python
#take all values from dict and convert to list to find the sum of list and get no of elements in dict to calculate the average

print(sum(list(price_dict.values()))/len(price_dict))
```

#Task 4 - Combined Operations

```python
catalog = []
for product, category in zip(products, categories):
  try:
    price = price_dict[product]
    catalog.append((product, price, category))
  except:
    pass       #if the product is not available in price_dict then pass, 'macbook' product was removed
```

```python
print(catalog)
```

```python
#for each category in catalog create a empty list and then traverse and add corresponding products
category_to_products ={}
for product,price,category in catalog:
  if category not in category_to_products:
    category_to_products[category] = []
  category_to_products[category].append(product)
```

```python
print(category_to_products)
```

```python
#find the max count
max_product_count = 0
for v in category_to_products.values():
  if max_product_count < len(v):
    max_product_count = len(v)

#the corresponding category having max count products is printed
for k,v in category_to_products.items():
  if len(v)==max_product_count:
    for p in v:
      print(p)
```
